```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
spendingData = pd.read_csv("Updated data with state spending and population.csv")
accidentsData = pd.read_csv("State_Information.csv")
#merged_data = pd.merge(accidentsData,spendingData,on=['State'])
#merged_data['Accidents Per Capita'] = (merged_data['Number Of Occurences'] / merged_data['State Population']) 
#merged_data['Spending Per Capita'] = merged_data['State Spending'] / merged_data['State Population']
#sns.scatterplot(x='Accidents Per Capita', y='Spending Per Capita', data=merged_data)
new_Data = accidentsData.copy()
new_Data['Accidents Per Capita'] = (accidentsData['Incidents'] / accidentsData['State Population'])
new_Data['Spending Per Capita'] = (accidentsData['State Spending'] / accidentsData['State Population'])
sns.scatterplot(x='Accidents Per Capita', y= 'Spending Per Capita', data=new_Data)
plt.title('Accidents per Capita vs. Spending per Capita')
plt.xlabel('Accidents Per Capita')
plt.ylabel('Spending Per Capita')
plt.show()

print(new_Data['Accidents Per Capita'].corr(new_Data['Spending Per Capita']))

print(spendingData['Incidents'].corr(spendingData['Pop Density']))
```
